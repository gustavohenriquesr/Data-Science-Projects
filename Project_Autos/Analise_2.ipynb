###Carregando Dataset

```python
# Creating Dataframe

df = pd.read_csv('autos.csv', encoding='utf-8')
df.head(10)
```

```python
df.columns.tolist()
```

###Número de veículos pertencentes a cada marca

```python
plt.figure(figsize=(12,10))
ax = sns.countplot(y='brand', data=df, palette='Reds_r')
ax.set_ylabel('Marca', fontsize=15)
ax.set_xlabel('Número de Veículos', fontsize=15)
plt.show()
```

###Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Getting gearbox types
gearbox_types = df['gearbox'].drop_duplicates().tolist()
print(gearbox_types)

# Getting vehicles types
vehicles_types = df['vehicleType'].drop_duplicates().tolist()
print(vehicles_types)
```

```python
plt.figure(figsize=(12,10))
ax = sns.barplot(x='vehicleType', y='price', hue='gearbox', data=df)
ax.set_xlabel('Tipos de Veículo')
ax.set_ylabel('Preço Médio')
plt.legend(title='Caixa de Câmbio')
plt.show()
```
