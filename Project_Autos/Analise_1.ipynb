###Carregando Dataset

```python
# Creating DataFrame

df = pd.read_csv('autos.csv', encoding='utf-8')
df.head(10)
```

```python
# Columns names
df.columns
```

###Distribuição de Valores com base no Ano de Registro 

```python
plt.figure(figsize=(16,12))
ax = sns.histplot(df['yearOfRegistration'],  kde=True, color='g', discrete=True)
ax.set_xlabel("Ano de Registro", fontsize=15)
ax.set_ylabel("Densidade", fontsize=15)
plt.show()
```

###Variação da Faixa de Preço pelo Tipo de Veículo

```python
plt.figure(figsize=(12,10))
ax = sns.boxplot(x=df['vehicleType'], y=df['price'])
ax.set_ylabel('Preço', fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
plt.show()
```

###Contagem total de veículos à venda conforme o tipo de veículo

```python
plt.figure(figsize=(12,10))
ax = sns.countplot(x=df['vehicleType'], palette='BuPu')
ax.set_ylabel('Total de Veículos para Venda', fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
i = 0
for p in ax.patches:
  ax.annotate(format(p.get_height(), '.1f'),
              (p.get_x() + p.get_width() / 2., p.get_height()),
              ha = 'center', va = 'center',
              xytext = (0, 9),
              textcoords = 'offset points')
plt.show()
```
