###Carregando Dataset

```python
# Creating Dataframe

df = pd.read_csv('autos.csv', encoding='utf-8')
df.head(10)
```

###Preço médio de veículo por tipo de combustível e tipo de caixa de câmbio

```python
plt.figure(figsize=(12,10))
ax = sns.barplot(x='fuelType',y='price',hue='gearbox', data=df)
ax.set_xlabel('Tipo de Combustível', fontsize=15)
ax.set_ylabel('Preço Médio', fontsize=15)
plt.legend(title='Caixa de Câmbio')
plt.show()
```

###Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
plt.figure(figsize=(12,10))
ax = sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', data=df, palette='Dark2')
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_ylabel('Potência Média', fontsize=15)
plt.legend(title='Caixa de Câmbio')
plt.show()
```
